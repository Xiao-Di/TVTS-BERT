```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
pems04_path = os.getcwd() + '/data/PEMS04/pems04.npz'
data = np.load(pems04_path)['data'] # (16992, 307, 3)
ts = data[:,:,0].squeeze().transpose() # (307, 16992)
ts.shape
```

```python
plt.plot(ts[0])
```

```python
ts[0].max()
```

```python
ts[0].min()
```

```python
num_roads = ts.shape[0]
raw = []
for i in range(num_roads):
    array = ts[i]
    max = array.max()
    min = array.min()
    normalized = ((array - min) / (max - min)).tolist()
    raw.append(normalized)
    
data_quarter = np.array(raw).reshape(-1, 72)
data_quarter[:5, :]
```
